2) Na imagem do passo 1, aplique o filtro convolucional de Gradiente 
2.1) um filtro de Sobel para a derivada na direção x, gerando uma imagem/matriz, A, com os valores da derivada em cada pixel/elemento da matriz;
2.2) um filtro de Sobel para a derivada na direção y, gerando uma imagens/matriz, B, com os valores da derivada em cada pixel/elemento da matriz;
2.3) em cada uma das matrizes, A e B, eleve ao quadrado os valores dos elementos;
2.4) some as duas matrizes A e B modificadas no passo 2.3 e tire a raiz quadrada  de cada elemento dessa matriz, C
3) Escolha um valor (float)  para threshold e 
4) gere uma matriz Final, D, com 
* pixel 0 caso o pixel correspondente da matriz C seja menor do que o threshold
* pixel 1. caso o pixel correspondente da matriz C seja maior do que o threshold.

```python
import numpy as np
from scipy.ndimage import gaussian_filter
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
from PIL import Image
img = []
img.append(Image.open(r"src\rika.png"))
img.append(Image.open(r"src\rika_cat.png"))
img.append(Image.open(r"src\neco_tr.png"))
img.append(Image.open(r"src\rena.jpg"))
img.append(Image.open(r"src\lambda_and_bern.jpg"))

def gaussian_blur(image, sigma=1):
    return gaussian_filter(image, sigma=sigma)

def sobel_filter(image):
    sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    
    gradient_x = convolve2d(image, sobel_x, mode='same', boundary='symm')
    gradient_y = convolve2d(image, sobel_y, mode='same', boundary='symm')
    
    gradient_x_squared = gradient_x ** 2
    gradient_y_squared = gradient_y ** 2
    
    gradient_magnitude = np.sqrt(gradient_x_squared + gradient_y_squared)
    
    return gradient_magnitude

def apply_threshold(image, threshold):
    return np.where(image < threshold, 0, 1)

def display_image(image):
    plt.imshow(image, cmap='binary')
    plt.axis('off')
    plt.show()

def random_image():
    # Exemplo de imagem
    return np.random.rand(100, 100) # Imagem aleatória de 100x100

def Alg01(image):
    # Passo 1: Suavização da imagem com filtro Gaussiano
    smoothed_image = gaussian_blur(image)
    
    # Passo 2: Aplicação do filtro convolucional de Gradiente
    gradient_image = sobel_filter(smoothed_image)
    
    # Passo 3: Escolha do valor de threshold
    threshold = 0.5
    
    # Passo 4: Geração da matriz final com base no threshold
    final_image = apply_threshold(gradient_image, threshold)
    
    # Exemplo de exibição da matriz final
    print("Matriz Final:")
    print(final_image)
    display_image(final_image)

#image = Image.open(r"src\rika.png")
#image = np.array(image)
#image = np.delete(image, range(0, image.shape[2]-1), axis=2)
#image = np.squeeze(image, axis=2)
#Alg01(image)

for i in range (5):
    image = np.array(img[i])
    image = np.delete(image, range(0, image.shape[2]-1), axis=2)
    image = np.delete(image, range(0, image.shape[2]-1), axis=2)
    image = np.squeeze(image, axis=2)
    Alg01(image)
```

Alg2.
1) Suavize a imagem, aplicando um filtro Gaussiano
2) Na imagem do passo 1, aplique o filtro convolucional de Laplace, gerando uma imagem/matriz A
3) Por simplicidade, gere uma imagem/matriz, B, percorrendo a imagem A e escrevendo em B
* 1. caso o pixel correspondente da matriz A seja diferente de 0 dentro de uma tolerância (0.0001, por exemplo)
* 0. cado o pixel correspondente da matriz A seja igual a 0 dentro de uma tolerância (0.0001, por exemplo).


```python
def laplace_filter(image):
    laplace_kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
    laplace_image = convolve2d(image, laplace_kernel, mode='same', boundary='symm')
    return laplace_image

def generate_binary_image(image, tolerance):
    binary_image = np.where(np.abs(image) > tolerance, 1, 0)
    return binary_image

def Alg02(image):
    # Passo 1: Suavização da imagem com filtro Gaussiano
    smoothed_image = gaussian_blur(image)
    
    # Passo 2: Aplicação do filtro convolucional de Laplace
    laplace_image = laplace_filter(smoothed_image)
    
    # Passo 3: Geração da imagem binária com base na tolerância
    binary_image = generate_binary_image(laplace_image, tolerance=1e-4)
    
    # Exemplo de exibição da matriz final
    print("Matriz Final:")
    print(binary_image)
    display_image(binary_image)

#image = Image.open(r"src\rika.png")
#image = np.array(image)
#image = np.delete(image, range(0, image.shape[2]-1), axis=2)
#image = np.squeeze(image, axis=2)
#Alg02(image)

for i in range (5):
    image = np.array(img[i])
    image = np.delete(image, range(0, image.shape[2]-1), axis=2)
    image = np.delete(image, range(0, image.shape[2]-1), axis=2)
    image = np.squeeze(image, axis=2)
    Alg02(image)
```
